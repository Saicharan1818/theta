```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data= {
    "Temperature": [22,55,65,98,42,15,36,45,78,98,66,45,14,7],
    "Humidity": [65,98,14,2,54,15,42,48,75,12,32,42,65,6],
    "wind speed" : [15,32,6,5,4,7,8,5,2,6,2,61,14,13],
    "pressure" : [1015,2302,1548,6547,4521,1452,2366,1547,1042,1025,1023,1564,1254,1425]

}
df=pd.DataFrame(data)
df

#scatter plots : we will get only few features
plt.figure(figsize=(10,6))
plt.scatter(df['Temperature'], df['Humidity'], color="blue", label="temp vs humidity")
plt.scatter(df['Temperature'], df['wind speed'], color="red", label="temp vs wind speed")
plt.scatter(df['Temperature'], df['pressure'], color="green", label="temp vs pressure")

#adding lables and title
plt.xlabel("Temperature")
plt.ylabel("all")
plt.title("Temperature vs Humidity", fontsize=14)
plt.legend()
plt.show()

#pair plot : all the features we will get
sns.pairplot(df)
plt.show()
```
